## Neural Network with hidden layer

Now let us create a NN with 5 input features and 2 layer where 1st layer contains 3 bias and second layer contains 1 bias and then it predicts the output. The hidden layer uses RelU and the output uses Sigmoid function.

```python
import torch
import torch.nn as nn
```

```python
class NeuralNetwork(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear1 = nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
```

```python
features = torch.rand(10, 5)

model = NeuralNetwork(features.shape[1])

model(features)
```

```python
model.linear2.weight
```

A NN can have multiple layers and passing the out variable through each layer eachtime is hectic hence we can use containers to make it much more simple.

```python
class NeuralNetwork2(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )


  def forward(self, features):
    out = self.network(features)

    return out
```

```python
model = NeuralNetwork2(features.shape[1])

model(features)
```

Now lets improve the training pipeline more to ensure that everything works fine

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/breast-cancer.csv')
```

```python
df.drop(columns=['id'], inplace=True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
```

```python
y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
```

```python
class NeuralNetwork3(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out

  def loss_function(self, y_pred, y):
    lepsilon = 1e-7
    y_pred = torch.clamp(y_pred, lepsilon, 1-lepsilon)
    loss = -(y * torch.log(y_pred) + (1 - y) *  torch.log(1 - y_pred)).mean()
    return loss
```

```python
model = NeuralNetwork3(X_train.shape[1])
```

```python
learning_rate = 0.01
num_epochs = 100

for epoch in range(num_epochs):
  y_pred = model(X_train)

  loss = model.loss_function(y_pred, y_train)

  loss.backward()

  with torch.no_grad():
    model.linear.weight -= learning_rate * model.linear.weight.grad
    model.linear.bias -= learning_rate * model.linear.bias.grad


  model.linear.weight.grad.zero_()
  model.linear.bias.grad.zero_()

  if epoch % 10 == 0:
    print(f'epoch: {epoch}, loss: {loss.item()}')


```

```python
with torch.no_grad():
  y_pred = model.forward(X_test)
  y_pred_class = y_pred.round()
  acc = accuracy_score(y_test, y_pred_class)
print(acc)
```

```python
class NeuralNetwork4(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
import torch.optim as opt
model = NeuralNetwork4(X_train.shape[1])
loss_function = nn.BCELoss()

optimiser = opt.SGD(model.parameters(), lr=learning_rate)
# optimiser
```

```python
learning_rate = 0.01
num_epochs = 100

for epoch in range(num_epochs):
  y_pred = model(X_train)

  loss = loss_function(y_pred, y_train.view(-1,1))


  # it is good to use the cleaning of gradients before the backward pass as it can make the working fluent and better
  optimiser.zero_grad()

  loss.backward()

  optimiser.step()


  if epoch % 10 == 0:
    print(f'epoch: {epoch}, loss: {loss.item()}')


```

```python
with torch.no_grad():
  y_pred = model.forward(X_test)
  y_pred_class = y_pred.round()
  acc = accuracy_score(y_test, y_pred_class)
print(acc)
```
